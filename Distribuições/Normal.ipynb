```python
import pandas as pd 
import seaborn as sns
import numpy as np 
from scipy import stats
import random
```

```python
# Gera uma distruibução normal de -3 a 3
dados_normal = stats.norm.rvs( size = 1000, random_state = 1)
```

```python
# Distribuição Normal
sns.displot(dados_normal, kde = True)
```

```python
dados_normal.mean, np.median(dados_normal), stats.mode(dados_normal), np.var(dados_normal), np.std(dados_normal)
```

```python
np.sum((dados_normal >= 0.9810041339322116) & (dados_normal <= 0.9810041339322116 + 1))
```

```python
np.sum((dados_normal <= 0.9810041339322116) & (dados_normal >= 0.9810041339322116 - 1))
```

```python
media = 160  # média desejada
desvio_padrao = 10  # desvio padrão desejado

# Gerar dados de altura com distribuição normal
dados_altura = np.random.normal(loc=media, scale=desvio_padrao, size=100)

# Arredondar os valores para uma casa decimal
dados_altura = np.round(dados_altura, 1)

# Se desejar garantir que todas as alturas estejam dentro de um intervalo específico
dados_altura = np.clip(dados_altura, 126, 192)

print(dados_altura)
```
